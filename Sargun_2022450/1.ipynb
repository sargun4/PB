```python
import requests

def run_sift_analysis(protein_sequence):
    url = "https://sift.bii.a-star.edu.sg/www/SIFT_seq_submit2.html"
    params = {
        "seq": protein_sequence,
        "siftSubmit": "Submit",
        "siftseqtxt": "",
        "siftupload": "",
    }
    response = requests.post(url, data=params)
    # Process the response and extract relevant information
    # (e.g., SIFT scores or predicted impact of variants)
    # In this example, we'll simply print the response text
    print(response.text)

# Define the normal and SNP-containing protein sequences
normal_protein_sequence = "MVHLTPVEKSAVTALWGKVNVDEVGGEALGRLLVVYPWTQRFFASFE"
snp_protein_sequence = "MVHLTPVEKSAVTALWGKVNVDEVGGEALGRLLVVYPWTQRFFASFE"  # Example SNP

# Run SIFT analysis for both sequences
print("Normal protein sequence:")
run_sift_analysis(normal_protein_sequence)

print("\nSNP-containing protein sequence:")
run_sift_analysis(snp_protein_sequence)
```
