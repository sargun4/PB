```python
import numpy as np

# Step 1: Create a base dataset
# For simplicity, let's consider a small dataset with 3 amino acid sequences
base_dataset = [
    "HEAGAWGHEE",
    "PAWHEAE",
    "MAAAWHEAE"
]

# Step 2: Define accepted point mutations
def accepted_point_mutation(gene, mutation_position, new_amino_acid):
    mutated_gene = list(gene)
    mutated_gene[mutation_position] = new_amino_acid
    return ''.join(mutated_gene)

# Example mutation
mutated_sequence = accepted_point_mutation(base_dataset[0], 3, 'C')
print(f"Mutated Sequence: {mutated_sequence}")

# Step 3: Calculate PAM rate of proteins
# For simplicity, let's assume a PAM1 matrix
PAM1_matrix = np.array([
    [0.987, 0.002, 0.001, ...],  # Row for Ala
    [0.001, 0.985, 0.003, ...],  # Row for Cys
    ...
])

# Step 4: Calculate Mutation Probability over 1 PAM
def mutation_probability_over_1_PAM(mutated_sequence, original_sequence, PAM_matrix):
    probability = 1.0
    for i in range(len(original_sequence)):
        original_aa = original_sequence[i]
        mutated_aa = mutated_sequence[i]
        probability *= PAM_matrix[original_aa][mutated_aa]
    return probability

# Example mutation probability
mutation_prob = mutation_probability_over_1_PAM(mutated_sequence, base_dataset[0], PAM1_matrix)
print(f"Mutation Probability over 1 PAM: {mutation_prob}")

# Step 5: Define PAM 250 matrix (Example)
PAM250_matrix = np.array([
    [1.5, -0.5, -2, ...],  # Row for Ala
    [-0.5, 2.0, 1, ...],   # Row for Cys
    ...
])

# Example usage of PAM 250 matrix
alignment_score = PAM250_matrix[original_aa][mutated_aa]
print(f"Alignment Score using PAM 250: {alignment_score}")
```
